```python
Perform the following operations on iris dataset
1. List down the features and their types
2. Create a box plot and histogram for each feature in the dataset.
3. Compare distributions and identify outliers.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target  # Adding target column

# 1. List down the features and their types
print("Features and their types:")
print(iris_df.dtypes)

# 2. Create a box plot and histogram for each feature in the dataset
# Box plot for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=iris_df, y=feature)
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()

# Histogram for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(iris_df[feature], bins=20, kde=True)
    plt.title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()

# 3. Compare distributions and identify outliers
# From the box plots and histograms, you can compare the distributions of each feature.
# Outliers can be identified by observing points that fall significantly outside the whiskers in the box plots.
```
