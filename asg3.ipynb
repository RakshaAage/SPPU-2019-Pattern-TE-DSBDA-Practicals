```python
import pandas as pd
import numpy as np
# Load the Iris dataset from CSV
iris_df = pd.read_csv("Iris.csv")

# 1. Display mean, median, minimum, maximum, standard deviation
print("Statistics for Iris dataset:")
print(iris_df.describe())

# 2. Provide mean, median, minimum, maximum, standard deviation for a given dataset by grouping using one of the qualitative (categorical) variable
# For example, let's group by the 'Species' variable
grouped_stats = iris_df.groupby('Species').describe()
print("\nStatistics for Iris dataset grouped by 'Species':")
print(grouped_stats)

# 3. Display missing values and inconsistencies
print("\nMissing Values and Inconsistencies:")
print(iris_df.isnull().sum())  # Check missing values

# 4. Replace missing values using any 2 suitable methods
# For example, we can replace missing values with mean and median
# Replace missing values in each column with mean and median respectively
# Exclude non-numeric columns before filling missing values
numeric_columns = iris_df.select_dtypes(include=np.number).columns
iris_df_filled_mean = iris_df.fillna(iris_df[numeric_columns].mean())
iris_df_filled_median = iris_df.fillna(iris_df[numeric_columns].median())


# Display modified DataFrame
print("\nModified DataFrame with missing values replaced using mean:")
print(iris_df_filled_mean.head())

print("\nModified DataFrame with missing values replaced using median:")
print(iris_df_filled_median.head())
```
